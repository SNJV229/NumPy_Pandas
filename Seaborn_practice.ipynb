## Style and colour:-

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# load the tips dataset present by default in seaborn
tips = sns.load_dataset('tips')
sns.set_style('white')    # This sets the style, we just need to replace white with below given themes

# The ways of styling themes are as follows:
    # white
    # dark
    # whitegrid
    # darkgrid
    # ticks
 
# make a countplot
sns.countplot(x ='sex', data = tips)
```

## Distribution plot:-

```python
import matplotlib.pyplot as plt
import seaborn as sns, numpy as np

x = np.random.randn(100)
ax = sns.distplot(x)
plt.show()
```

## Categorical plot:-

```python
import seaborn as sns
%matplotlib inline 
t=sns.load_dataset('tips')
sns.barplot(x='sex',y='total_bill',data=t)
```

```python
sns.countplot(x='sex',data=t)
```

```python
sns.boxplot(x='day',y='total_bill',data=t,palette='rainbow')
```

```python
import seaborn as sns
%matplotlib inline

flights= sns.load_dataset('flights')
fp= flights.pivot_table(index='month', columns='year',values='passengers')  

sns.heatmap(fp)  # we can use cmap='lukewarm' or 'coolwarm' as another parameter for different pattern in the plot
```

## Grid:-

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(data = iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)        # Plot histogram on digonals of the grid.
g.map_upper(plt.scatter)    # Plot scatter on upper part of dagonals of the grid.
g.map_lower(sns.kdeplot)    # Plot kdeplot on lower part of digonals of the grid.
```
